```python
import pandas as pd

file_path = '/content/Scientific LLM Questions Bank - Questions.csv'
questions_df = pd.read_csv(file_path)

questions_df.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
import sacrebleu

def calculate_similarity(ideal_answer, generated_answer):
    ideal_answer = str(ideal_answer)
    generated_answer = str(generated_answer)
    vectorizer = TfidfVectorizer()
    vectors = vectorizer.fit_transform([ideal_answer, generated_answer])
    similarity = cosine_similarity(vectors[0], vectors[1])[0][0]
    return similarity

def calculate_bleu(ideal_answer, generated_answer):
    ideal_answer = str(ideal_answer)
    generated_answer = str(generated_answer)
    bleu = sacrebleu.sentence_bleu(generated_answer, [ideal_answer])
    return bleu.score

model_columns = ["EleutherAI/pythia-410m",	"openai-community/gpt2",	"EleutherAI/pythia-410m (RAG)",	"openai-community/gpt2 (RAG)",	"EleutherAI/pythia-410m (Finetuned 35 examples)",	"EleutherAI/pythia-410m (Finetuned 150 examples)",	"Human Subject A (19 yo female history major)",	"Human Subject B (21 yo female biology major)",	"Human Subject C (20 yo Male agricultural major)"]

for col in model_columns:
    questions_df[f'{col} Similarity'] = questions_df.apply(lambda row: calculate_similarity(row['Ideal Answer'], row[col]), axis=1)
    questions_df[f'{col} BLEU'] = questions_df.apply(lambda row: calculate_bleu(row['Ideal Answer'], row[col]), axis=1)
```

```python
averages_similarity = questions_df[['EleutherAI/pythia-410m Similarity', 'openai-community/gpt2 Similarity', 'EleutherAI/pythia-410m (RAG) Similarity', 'openai-community/gpt2 (RAG) Similarity', "EleutherAI/pythia-410m (Finetuned 35 examples) Similarity", "EleutherAI/pythia-410m (Finetuned 150 examples) Similarity", "Human Subject A (19 yo female history major) Similarity","Human Subject B (21 yo female biology major) Similarity", "Human Subject C (20 yo Male agricultural major) Similarity"]].mean()
print("\nAverages for Similarity Columns:")
print(averages_similarity)

averages_bleu = questions_df[['EleutherAI/pythia-410m BLEU', 'openai-community/gpt2 BLEU', 'EleutherAI/pythia-410m (RAG) BLEU', 'openai-community/gpt2 (RAG) BLEU', "EleutherAI/pythia-410m (Finetuned 35 examples) BLEU", "EleutherAI/pythia-410m (Finetuned 150 examples) BLEU", "Human Subject A (19 yo female history major) BLEU", "Human Subject B (21 yo female biology major) BLEU", "Human Subject C (20 yo Male agricultural major) BLEU"]].mean()
print("\nAverages for BLEU Columns:")
print(averages_bleu)
```
